```python
# 2026-03-02 run against production
from tqdm import tqdm
from apis_core.apis_entities.models import Person
```

```python
items = Person.objects.filter(name__startswith="von ")
items.count()
```

```python
for x in tqdm(items):
    x.name = x.name[4:]
    x.save()
```
