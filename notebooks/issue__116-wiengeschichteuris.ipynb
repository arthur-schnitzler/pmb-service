```python
# 2026-06-03
from apis_core.apis_metainfo.models import Uri
from django.core.exceptions import ObjectDoesNotExist
from csae_pyutils import gsheet_to_df
from apis_core.apis_metainfo.models import TempEntityClass
from apis_core.utils import get_object_from_pk_or_uri

from tqdm import tqdm
```

```python
domain = "wiengeschichtewiki"
```

```python
df = gsheet_to_df("1dXjEymTuoaXKMyX9OMyuXE7snP9xQzZdNdG4ZFVfbs8")
```

```python
df.head()
```

```python
for i, row in tqdm(df.iterrows()):
    try:
        entity = TempEntityClass.objects.get(id=row["pmb_id"])
    except ObjectDoesNotExist:
        entity = get_object_from_pk_or_uri(row["pmb_id"])
    wr_uri = f"https://www.geschichtewiki.wien.gv.at/Special:URIResolver/?curid={row['wiengeschichte_id']}"
    try:
        uri = Uri.objects.get_or_create(uri=wr_uri, domain=domain, entity=entity)
    except Exception as e:
        print(row["pmb_id"], wr_uri, e)
```
