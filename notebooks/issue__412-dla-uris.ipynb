```python
# 2026-04-08
from tqdm import tqdm
from csae_pyutils import gsheet_to_df
from apis_core.apis_metainfo.models import Collection, CollectionType, Uri, TempEntityClass
```

```python
df = gsheet_to_df("13ztmIBfB2q7bS7Ur8D66hKG3lV43GN0vju0wczUaHeg")
```

```python
col, _ = Collection.objects.get_or_create(name="DLA-Schnitzler")
domain = "dla-marbach"
col_type, _ = CollectionType.objects.get_or_create(name="Projekt")
col.description = 'To be done'
col.collection_type = col_type
col.save()
```

```python
does_not_exist = []
for i, row in tqdm(df.iterrows()):
    tmp_uri = row["dla_url"]
    ent_id = row["pmb_id"]
    try:
        entity = TempEntityClass.objects.get(id=ent_id)
    except Exception as e:
        does_not_exist.append([ent_id, e])
    if entity:
        pmb_uri, _ = Uri.objects.get_or_create(uri=tmp_uri, domain=domain)
        pmb_uri.entity = entity
        pmb_uri.save()
        entity.collection.add(col)
```
