```python
# 2026-03-02 run against production
from tqdm import tqdm
from apis_core.apis_metainfo.models import Uri
```

```python
uris = Uri.objects.filter(domain="wedekind-korrespondenz")
```

```python
uris.count()
```

```python
uris.first().uri
```

```python
for x in tqdm(uris, total=uris.count()):
    new_uri = x.uri.replace("/list.xhtml?", "/single.xhtml?")
    x.uri = new_uri
    x.save()
```
