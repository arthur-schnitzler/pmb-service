```python
# 2025-12-06 run against production
from tqdm import tqdm
```

```python
base_url = "https://www.kraus.wienbibliothek.at/legalkraus/"
uris = Uri.objects.filter(uri__startswith="https://kraus.wienbibliothek.at/register/")
```

```python
uris.count()
```

```python
for x in tqdm(uris, total=uris.count()):
    pmb_id = x.uri.split("/")[-1]
    new_uri = f"{base_url}{pmb_id}.html"
    x.uri = new_uri
    x.save()
```
